```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('csgo (1).csv')
```

```python
le = LabelEncoder()

```

```python
df['bomb_planted'] = le.fit_transform(df['bomb_planted'])
df['map'] = le.fit_transform(df['map'])
df['round_winner'] = le.fit_transform(df['round_winner'])
```

```python
df.info()
```

```python
x = df.iloc[:,:-1]
y = df['round_winner']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=34)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
log_pred = log_model.predict(x_test)
```

```python
log_pred
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test,log_pred)
```

```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)

lda_coefficient = np.exp(np.abs(lda.coef_))
```

```python
lda_coefficient
```

```python
lda_coefficient = lda_coefficient.flatten()
```

```python
lda_coefficient
```

```python
feature_names = x.columns
```

```python
feature_names
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.bar(feature_names,lda_coefficient)
plt.xticks(rotation = 90)
plt.xlabel('Features')
plt.ylabel('Score')
plt.show()
```

```python
df_features_score = pd.DataFrame({"Feature Names": feature_names, "Feature Score":lda_coefficient })
```

```python
df_features_score
```

```python
top_20_values = df_features_score.nlargest(20,'Feature Score')
top_20_values
```
